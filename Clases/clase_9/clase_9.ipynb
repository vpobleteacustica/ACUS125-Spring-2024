> + Escribir las ecuaciones de movimiento:
$$

$$

> + Esta ecuaciones representan un sistema de ecuaciones diferenciales **acopladas**, es significa físicamente que el movimiento de $x_{1}$ afecta a $x_{2}$ y viceversa.

> + Condiciones iniciales: Para las masas $m_{1}$ y $m_{2}$, tenemos los desplazamientos y velocidades iniciales: $x_{1}(0) = x_{10},$ $\dot{x}_{1}(0) = v_{10}$; $x_{2}(0) = x_{20}$, y $\dot{x}_{2}(0) = v_{20}.$

> + Escribir las ecuaciones en forma matricial:
$$
\ldots
$$

> + Lo cual se puede re-escribir como:

$$
\begin{align}
\text{M}{\ddot{{\bf{x}}}} + \text{K}{\bf{x}} &={\bf{0}}
\end{align}
$$

> + donde ${\ddot{{\bf{x}}}}=\left[ \ddot{x}_{1}\,\,\ddot{x}_{2}\right]^{\text{T}}$; ${\bf{x}}(t)=\left[ x_{1}\,\,x_{2}\right]^{\text{T}}$ y ${\bf{0}}=\left[ 0\,\,0\right]^{\text{T}}.$ Las matrices $\text{M}$ y $\text{K}$ se denominan matrices de masa y de rigidez respectivamente.

> + Las matrices $\text{M}$ y $\text{K}$ son simétricas, es decir, $\text{M}=\text{M}^{\text{T}}$ y $\text{K}=\text{K}^{\text{T}}$ 

> + Asumir una solución armónica: 

$$
\begin{align}
{\bf{x}}(t) & = {\bf{u}}e^{j\omega t}
\end{align}
$$

> + donde ${\bf{u}}$ es un vector de desplazamiento: ${\bf{u}} = \left[ u_{1}\,\, u_{2}\right]^{T}$ que se debe determinar y además, debe ser distinto de cero.
> + $\omega$ también debe ser determinada.
> + Derivando la Ec. (2) y reemplazando en Ec. (1) se obtiene:

$$
\begin{align*}
\left( -\omega^{2}\text{M} + \text{K}\right)\cdot{\bf{u}}e^{j\omega t} &= {\bf{0}}
\end{align*}
$$
> + Omitimos la exponencial compleja $e^{j\omega t}$ porque siempre es distinto de cero para todo $t.$
> + El vector de desplazamiento ${\bf{u}}$ también debe ser distinto de cero.
> + Para que el sistema de ecuaciones tenga solución no trivial, la matriz $\left( -\omega^{2}\text{M} + \text{K}\right)$ no debe tener inversa, porque si la tuviese, se podría pre-multiplicar por $\left( -\omega^{2}\text{M} + \text{K}\right)^{-1}$ y se obtendría ${\bf{u}}=0.$
> + La condición que la matriz $\left( -\omega^{2}\text{M} + \text{K}\right)$ no tenga inversa es equivalente a expresar que su determinante sea cero:

$$
\text{det}\,\begin{bmatrix}
-\omega^{2}m_{1}+k_{1}+k_{2} & -k_{2}\\
-k_{2} & -\omega^{2}m_{2}k_{2}
\end{bmatrix}=0
$$

> + de donde se obtiene que $\omega^{2}$ debe satisfacer:

$$
\begin{align*}
m_{1}m_{2}\omega^{4}-(m_{1}k_{2}+m_{2}k_{1}+m_{2}k_{2})+k_{1}k_{2}=0
\end{align*}
$$

> + Esta expresión se llama ecuación característica del sistema.

> + Para cada valor de $\omega^{2}$ (es decir, $\omega_{1}^{2}$ y $\omega_{2}^{2}$), se escribe un vector ${\bf{u}}$ que satisface:

$$
\begin{align*}
\left( -\omega_{1}^{2}\text{M} + \text{K}\right)\cdot{\bf{u}}_{1}&= {\bf{0}}
\end{align*}
$$
$$
\begin{align*}
\left( -\omega_{2}^{2}\text{M} + \text{K}\right)\cdot{\bf{u}}_{2}&= {\bf{0}}
\end{align*}
$$

> + Con estas expresiones obtenemos la dirección de los vectores ${\bf{u}}_{1}$ y ${\bf{u}}_{2}$ pero no su magnitud por un tema de escalamiento. Es decir, si ${\bf{u}}_{1}$ es una solución, entonces también lo es: $a{\bf{u}}_{1}$ donde $a$ es una constante distinta de cero.

> + Por lo tanto, los vectores ${\bf{u}}_{1}$ y ${\bf{u}}_{2}$ tienen magnitud arbitraria.

> + la solución general es una combinación lineal de:

$$
\begin{align*}
{\bf{x}}(t) & = \left( ae^{-j\omega_{1} t} + be^{j\omega_{1} t} \right){\bf{u}}_{1} +  \left( ce^{-j\omega_{2} t} + de^{j\omega_{2} t} \right){\bf{u}}_{2}
\end{align*}
$$

> + donde $a$, $b$, $c$ y $d$ son constantes a determinar de las condiciones iniciales.

> + También, se puede re-escribir como:

$$
\begin{align*}
{\bf{x}}(t) & = A_{1}\sin(\omega_{1}t+\phi_{1})\, {\bf{u}}_{1}+ A_{2}\sin(\omega_{2}t+\phi_{2})\, {\bf{u}}_{2}
\end{align*}
$$

> + Aquí se revela uno de los puntos más importantes del curso:

$$
\begin{align*}
\begin{bmatrix}
x_{1}\\
x_{2}
\end{bmatrix}&= {\underbrace{\begin{bmatrix}
\begin{Bmatrix}
u_{11} \\
u_{21} 
\end{Bmatrix} & \begin{Bmatrix}
u_{12} \\
u_{22} 
\end{Bmatrix}
\end{bmatrix}}_{\text{matriz de modos $\left[P\right]=\left[\left\{u_{1}\right\}\,\, \left\{u_{2}\right\}\right] $}}}\cdot \begin{bmatrix}
A_{1}\sin(\omega_{1}t+\phi_{1}) \\
A_{2}\sin(\omega_{2}t+\phi_{2}) 
\end{bmatrix}
\end{align*}
$$

> + Esto quiere decir, que cada masa oscila con dos frecuencias $\omega_{1}$ y $\omega_{2}.$ Estas se ordenan siempre de menor a mayor.

> + Estas frecuencias se denominan frecuencias naturales del sistema.

> + Si $A_{2}=0$, entonces cada masa oscilará a la frecuencia natural $\omega_{1}$ y la posición relativa de las masas en todo instante, va a estar dada por el vector modo de vibrar ${\bf{u}}_{1}.$ Este vector será denominado el primer modo del sistema.

> + Análogamente, si $A_{1}=0$, las masas oscilan a la frecuencia natural $\omega_{2}$ de acuerdo al vector modo de vibrar ${\bf{u}}_{2}.$

> + Este segundo vector se llama segundo modo del sistema.

### Valores propios y frecuencias naturales:

> + Asumir que tenemos un sistema con múltiples grados de libertad: 

$$
\begin{align}
\text{M}{\ddot{{\bf{x}}}} + \text{K}{\bf{x}} &={\bf{0}}
\end{align}
$$

> + Pre-multipliquemos por $\text{M}^{-1}:$

$$
\begin{align}
{\underbrace{\text{M}^{-1}\text{M}}_{\text{I}}}\,{\ddot{{\bf{x}}}} + \text{M}^{-1}\text{K}\,{\bf{x}} &={\bf{0}}\\
\text{I}\,{\ddot{{\bf{x}}}}+\text{M}^{-1}\text{K}\,{\bf{x}} &={\bf{0}} 
\end{align}
$$

> + Si suponemos una solución de la forma ${\bf{x}}(t) = {\bf{u}}e^{j\omega t}$

$$
\begin{align}
\left( \text{M}^{-1}\text{K}\, -\omega^{2}\text{I}\right)\,{\bf{u}} &={\bf{0}}
\end{align}
$$

> + Llamemos a un escalar $\lambda = \omega^{2}\text{I}$ y a una matriz $A=\text{M}^{-1}\text{K}$, entonces obtenemos:

$$
\begin{align*}
A{\bf{u}}&=\lambda {\bf{u}}
\end{align*}
$$

> + Esto es un problema de valores propios!

> + El escalar $\lambda$ se llama valor propio y es la frecuencia natural del sistema y el vector ${\bf{u}}$ correspondiente se llama vector propio del sistema.

> + Luego, podemos decir que las frecuencias naturales y modos de vibrar se determinan a partir de los valores propios y vectores propios, respectivamente, de la matriz $\text{M}^{-1}\text{K}$.

> + Una propiedad importante de los vectores propios es que son **ortogonales**.

> + Esto significa que **el producto punto entre dos vectores es igual a cero**:

$$
\begin{align*}
{\bf{u}}_{1}^{T}{\bf{u}}_{2} &=0
\end{align*}
$$

> + Aunque los vectores propios tienen una magnitud arbitraria, siempre se pueden normalizar (dividir por algo!) para tener largo unitario.

> + Así, podemos determinar vectores propios **ortonormales**.

> + La norma o magnitud de un vector es: 

$$
\begin{align*}
\left\Vert{\bf{x}}\right\Vert &= \sqrt{{\bf{x}}^{T}{\bf{x}}}=\left[ \sum_{i}x_{i}^{2}\right]^{1/2}
\end{align*}
$$

> + Este hecho de obtener vectores propios ortonormales **permite desacoplar las ecuaciones de movimiento**.

> + Supongamos que analizamos el caso sin amortiguamiento. Entonces, tenemos:

$$
\begin{align*}
\left(-\omega_{i}^{2}\text{M}+\text{K}\right)\,{\bf{u}}_{i} &=0
\end{align*}
$$

> + Consideremos dos modos de vibrar arbitrarios: $r$ y $s$

$$
\begin{align*}
\text{K}{\bf{u}}_{r} &= \omega_{r}^{2}\text{M}{\bf{u}}_{r}\\
\text{K}{\bf{u}}_{s} &= \omega_{s}^{2}\text{M}{\bf{u}}_{s}
\end{align*}
$$

> + Al pre-multiplicar la primera por ${\bf{u}}_{s}^{T}$ y la segunda por ${\bf{u}}_{r}^{T}$:

$$
\begin{align*}
{\bf{u}}_{s}^{T}\text{K}{\bf{u}}_{r} &= \omega_{r}^{2}{\bf{u}}_{s}^{T}\text{M}{\bf{u}}_{r}\\
{\bf{u}}_{r}^{T}\text{K}{\bf{u}}_{s} &= \omega_{s}^{2}{\bf{u}}_{r}^{T}\text{M}{\bf{u}}_{s}
\end{align*}
$$

> + La transpuesta del producto de matrices es $(A\cdot B)^{T} = B^{T}\cdot A^{T}$.

> + La transpuesta de la transpuesta de $A$ es $A$: $(A^{T})^{T} = A$.

> + La transpuesta del producto de un escalar $\alpha$ por una matriz $A$ es $(\alpha \cdot A)^{T} = \alpha\cdot A^{T}$

> + Si se transpone la segunda ecuación y se resta con la primera, obtenemos:

$$
\begin{align*}
\left( {\bf{u}}_{r}^{T}\text{K}{\bf{u}}_{s}\right)^{T} &= \left(\omega_{s}^{2}{\bf{u}}_{r}^{T}\text{M}{\bf{u}}_{s}\right)^{T}
\end{align*}
$$

$$
\begin{align*}
{\bf{u}}_{s}^{T}\text{K}{\bf{u}}_{r} &= \omega_{s}^{2}{\bf{u}}_{s}^{T}\text{M}{\bf{u}}_{r}
\end{align*}
$$

> + Al restar:

$$
\begin{align*}
0 &= \left( \omega_{r}^{2} - \omega_{s}^{2}\right) {\bf{u}}_{s}^{T}\text{M}{\bf{u}}_{r}
\end{align*}
$$

> + Esta ecuación nos da dos posibles resultados diferentes:

$$
\begin{align*}
{\bf{u}}_{s}^{T}\text{M}{\bf{u}}_{r} &= 0\quad r\neq s
\end{align*}
$$
$$
\begin{align*}
{\bf{u}}_{s}^{T}\text{M}{\bf{u}}_{r} \neq 0\quad r= s
\end{align*}
$$

> + La primera ecuación dice que dos modos distintos son ortogonales con la matriz
de masa. 

> + Y que cuando se calcula el producto del mismo modo con la matriz de masa como en la segunda ecuación, el producto no es cero. 

> + El valor de este producto se denomina **masa generalizada** $m_{ii}$:

$$
\begin{align*}
m_{ii}&={\bf{u}}_{i}^{T}\text{M}{\bf{u}}_{i}
\end{align*}
$$

> + De manera similar se define la **rigidez generalizada** $k_{ii}$:

$$
\begin{align*}
k_{ii}&={\bf{u}}_{i}^{T}\text{K}{\bf{u}}_{i}
\end{align*}
$$

> + Dado esto, obtenemos una relación importante:

$$
\begin{align*}
P^{T}MP&={\bf{m}}\\
P^{T}KP&={\bf{k}}\\
\end{align*}
$$

> + donde $P = \left[ {\bf{u}}_{1}\,\, {\bf{u}}_{2}\,\,\ldots {\bf{u}}_{n} \right]$ se denomina matriz de modos normales, ${\bf{m}}$ y ${\bf{k}}$ son matrices diagonales llamadas **matriz de masa modal** y **matriz de rigidez modal**, respectivamente.

> + Ya que la matriz de modos está sujeta a una normalización, los valores de $k_{ii}$ y $m_{ii}$, no son únicos. 

> + Sin embargo, si se puede demostrar que la razón $k_{ii}/m_{ii}$ es **única** y tiene un valor igual a $\omega_{i}^{2}$. 

### La matriz de modos:

> + La importancia de esta matriz, es que se usa para definir una **transformación de coordenadas.** 

> + Para ésto, se introduce una nueva coordenada ${\bf{y}}$, a veces llamada coordenada **Principal**.

> + La transformación de coordenadas es ${\bf{x}}(t)=P{\bf{y}}(t).$

> * Usando la Ec. (1) y pre-multiplicándola por $P^{T}$, obtenemos:

$$
\begin{align*}
{\underbrace{P^{T}MP}_{{\bf{m}}}}\,\,{\ddot{{\bf{y}}}} + {\underbrace{P^{T}KP}_{{\bf{k}}}}\,\,{\bf{y}} &={\bf{0}}
\end{align*}
$$

> + Para el caso de dos grados de libertad, veamos los cálculos en representación matricial :

$$
\begin{align*}
\begin{bmatrix}
m_{11} & 0\\
0      & m_{22}
\end{bmatrix}\cdot \begin{bmatrix}
\ddot{y}_{1} \\
\ddot{y}_{2}      
\end{bmatrix} +  \begin{bmatrix}
k_{11} & 0\\
0      & k_{22}
\end{bmatrix}\cdot \begin{bmatrix}
y_{1} \\
y_{2}      
\end{bmatrix} & = \begin{bmatrix}
0 \\
0      
\end{bmatrix}
\end{align*}
$$

$$
\begin{align*}
\begin{bmatrix}
m_{11}\ddot{y}_{1} + k_{11}y_{1} \\
m_{22}\ddot{y}_{2} + k_{22}y_{2}
\end{bmatrix} & = \begin{bmatrix}
0 \\
0      
\end{bmatrix}
\end{align*}
$$

> + Se observa el desacople de las ecuaciones:

$$
\begin{align*}
\ddot{y}_{1} + {\underbrace{\frac{k_{11}}{m_{11}}}_{\omega_{1}^{2}}}\,y_{1} &= 0 \end{align*}
$$

$$
\begin{align*}
\ddot{y}_{2} + {\underbrace{\frac{k_{22}}{m_{22}}}_{\omega_{2}^{2}}}\,y_{2} &= 0 \\
\end{align*}
$$

$$
\begin{align*}
\ddot{y}_{1} + \omega_{1}^{2}\,y_{1} &= 0 
\end{align*}
$$

$$
\begin{align*}
\ddot{y}_{2} + \omega_{2}^{2}\,y_{2} &= 0 
\end{align*}
$$

> + Estas ecuaciones son desacopladas porque cada una de ellas depende solamente de una variable.

> + Esto significa, que se resuelven de manera independiente.

> + Se les suele llamar **ecuaciones modales** y al sistema de coordenadas ${\bf{y}}=\left[ y_{1}\,\,y_{2}\right]$ se le da el nombre de **sistema de coordenadas modales.**

> + Estas nuevas ecuaciones deben cumplir con condiciones iniciales, las que se deben transformar al sistema de coordenadas principales $y(t):$

$$
\begin{align*}
{\bf{y}}_{0} &= P^{T}\,{\bf{x}}_{0} 
\end{align*}
$$

$$
\begin{align*}
\dot{{\bf{y}}}_{0} &= P^{T}\,{\bf{v}}_{0}
\end{align*}
$$

> + Definimos los vectores normales con magnitud unitaria, de tal forma que:

$$
\begin{align*}
P\cdot P^{T} &= I 
\end{align*}
$$

> + Y, por lo tanto,
$$
\begin{align*}
P^{-1} & = P^{T} 
\end{align*}
$$

> + La solución de cada ecuación es:

$$
\begin{align*}
y_{1}(t) = \frac{\sqrt{\omega_{1}^{2}y_{10}^{2} + \dot{y}_{10}^{2} }}{\omega_{1}}\cdot \sin\left( \omega_{1}(t) + \tan^{-1}\frac{\omega_{1}y_{10}}{\dot{y}_{10}}\right)
\end{align*}
$$

$$
\begin{align*}
y_{2}(t) = \frac{\sqrt{\omega_{2}^{2}y_{20}^{2} + \dot{y}_{10}^{2} }}{\omega_{2}}\cdot \sin\left( \omega_{2}(t) + \tan^{-1}\frac{\omega_{2}y_{20}}{\dot{y}_{20}}\right)
\end{align*}
$$

> + Cuando se obtengan estas soluciones para cada una de las ecuaciones, la solución general será:

$$
\begin{align*}
x(t) & = P\,y(t)\\
     & = \sum_{i=1}^{n} {\bf{u}}_{i}y_{i}(t)\\
     & = \sum_{i=1}^{n} {\bf{u}}_{i}d_{i}\sin\left( \omega_{i}(t) + \phi_{i}\right)
\end{align*}
$$

> + donde $d_{i}$ depende de las condiciones iniciales, que indicará cuánto pesa o contribuye cada uno de los modos en la respuesta total del sistema.

> + Si el modo $d_{i}$ es grande quiere decir que ese modo $i-$ésimo, contribuye de manera importante en la respuesta.

## Un sistema de dos masas con resortes

Este ejercicio demuestra la respuesta de vibración libre y el problema de valor propio/vector propio, usando un sistema de dos masas con resortes, que se muestra en la Figura 4. Sólo ilustraremos un conjunto de parámetros de ejemplo. Pero, las mismas técnicas se aplican para otros parámetros y para matrices más grados de libertad.

Escribimos las ecuaciones de movimiento del sistema:

$ \quad m_1 \ddot{x}_1 + (k_1+k_2)x_1 - k_2 x_2 = 0 $

$ \quad m_2 \ddot{x}_2 - k_2 x_1 + (k_2 + k_3)x_2 = 0 $

	<img src="eje.png" alt="MDOFCT" width=30%/></a><br>

También, podemos escribir estas ecuaciones en forma matricial:

$ \quad \begin{bmatrix}m_1 & 0 \\ 0 & m_2\end{bmatrix}\begin{bmatrix}\ddot{x}_1 \\ \ddot{x}_2\end{bmatrix} +  \begin{bmatrix}k_1 + k_2 & -k_2 \\ -k_2 & k_2 + k_3\end{bmatrix}\begin{bmatrix}x_1 \\ x_2\end{bmatrix} = \begin{bmatrix}0 \\ 0\end{bmatrix}$

Definimos las matrices:

$ \quad M = \begin{bmatrix}m_1 & 0 \\ 0 & m_2\end{bmatrix} $

y 

$ \quad K = \begin{bmatrix}k_1 + k_2 & -k_2 \\ -k_2 & k_2 + k_3\end{bmatrix} $

Usaremos la librería [Scipy versión del módulo de álgebra lineal (http://docs.scipy.org/doc/scipy-0.13.0/reference/generated/scipy.linalg.eigh.html). Esta librería es útil para solucionar el problema de valores propios en genral.


```python
from scipy import linalg
import numpy as np                        

from scipy.integrate import odeint        # importamos "odeint" para hacer simulaciones
from scipy import linalg                  # Usamos "linalg" para abordar el problema de valores propios

# Para no mostrar los plots en una ventana separada (displayed inline)
%matplotlib inline                        
import matplotlib.pyplot as plt
```

```python
def ecuacion_de_movimiento(w, t, p):
    """
    Ecuaciones diferenciales para el sistema de masas y resortes acopladas.

    Argumentos:
        w :  vector con las variables de estado:
                  w = [x1, x1_dot, x2, x2_dot]
        t :  tiempo
        p :  vector de parámetros:
                  p = [m1, m2, k1, k2, k3]
    """
    x1, x1_dot, x2, x2_dot = w
    m1, m2, k1, k2, k3 = p

    # Creamos sysODE = (x1', x1_dot', x2', x2_dot'):
    sysODE = [x1_dot,
             (-(k1 + k2) * x1 + k2 * x2) / m1,
             x2_dot,
             (k2 * x1 - (k2 + k3) * x2) / m2]
              
    return sysODE
```

```python
# Parámetros de nuestro sistema:
m1 = 1.0                # kg
m2 = 1.0                # kg
k1 = 0.4*10**6 # N/m
k2 = 0.8*10**6 # N/m
k3 = 0.4*10**6 # N/m
```

### Problema de valores propios

Usamos $M$ y $K$ para resolver:

$ \quad \left[- \omega^2 M + K\right]\vec{u} = 0 $ 

para $\vec{u}$.

```python
# Matrices
M = np.asarray([[m1, 0],
                [0,  m2]])

K = np.asarray([[k1 + k2, -k2],
                [-k2,      k2 + k3]])
```

```python
eigenvals, eigenvects = linalg.eigh(K, M)
eigenvals
```

La función linalg.eigh devuelve dos matrices, uno de los valores propios y otro de los vectores propios. Los valores propios son el cuadrado de las dos frecuencias naturales. Los vectores propios se devuelven en forma normalizada, donde cada ''fila'' de la matriz representa un vector propio.

```python
print('\n')
print('Los valores propios resultantes son {:.2f} and {:.2f}.'.format(eigenvals[0], eigenvals[1]))
print('\n')
print('Entonces las dos frecuencias naturales son {:.2f}rad/s and {:.2f}rad/s.'.format(np.sqrt(eigenvals[0]), np.sqrt(eigenvals[1])))
print('\n')
```
